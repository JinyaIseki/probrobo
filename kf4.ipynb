```python
import sys
sys.path.append('../scripts/')
from robot import *##scriptsの中のロボットや地図を作成するための関数を読み込む
from scipy.stats import multivariate_normal##多変量ガウス分布のモジュール
from matplotlib.patches import Ellipse##楕円描画用
```

```python
def sigma_ellipse(p, cov, n): ##楕円描画
    eig_vals, eig_vec = np.linalg.eig(cov)##covの固有値と固有ベクトルをeig_vals, eig_vecに代入
    ang = math.atan2(eig_vec[:,0][1], eig_vec[:,0][0])/math.pi*180#eig_vals, eig_vecから傾きを計算
    return Ellipse(p, width=n*math.sqrt(eig_vals[0]),height=n*math.sqrt(eig_vals[1]), angle=ang, fill=False, color="blue", alpha=0.5)#nｘσ範囲の楕円を描画
```

```python
class KalmanFilter: ###kf4init
    def __init__(self, envmap, init_pose, motion_noise_stds, distance_dev_rate=0.14, direction_dev=0.05): #変数追加
        self.belief = multivariate_normal(mean=init_pose, \
                                        cov=np.diag([1e-10, 1e-10, 1e-10])) 
        self.motion_noise_stds = motion_noise_stds
        self.map = envmap  #以下3行追加（Mclと同じ）
        self.distance_dev_rate = distance_dev_rate
        self.direction_dev = direction_dev
        
    #計測更新
    def observation_update(self, observation):  
        for d in observation:#観測できたランドマークごとに更新する
            z = d[0]
            obs_id = d[1]
            
            ###Hの計算###（h(x)をテイラー展開で一次関数に近似すると出てくる項）h(x)は確率変数がｚ（センサーの値）の確率密度関数の平均値
            mx, my = self.map.landmarks[obs_id].pos##計測したランドマークのx,y座標をmx,myに代入
            mux, muy, mut = self.belief.mean##信念の平均値（x,y,z）
            q = (mux - mx)**2 + (muy - my)**2
            sqrtq = np.sqrt(q)#二乗平均（割り算を省いている）
            H = np.array([[(mux - mx)/sqrtq, (muy - my)/sqrtq, 0.0],  [(my - muy)/q, (mux - mx)/q, -1.0]])
            
            ###Qの計算###Qは確率変数がｚ（センサーの値）の確率密度関数の分散
            hmu = IdealCamera.relative_polar_pos(self.belief.mean, self.map.landmarks[obs_id].pos)#信念と計測したランドマークの位置をhmuに代入
            distance_dev = self.distance_dev_rate*hmu[0]# 
            Q = np.diag(np.array([distance_dev**2, self.direction_dev**2]))#μ周りに共分散に近似したあとの式
            
            ###カルマンゲインの計算###
            K = self.belief.cov.dot(H.T).dot(np.linalg.inv(Q + H.dot(self.belief.cov).dot(H.T)))
            
            ###更新###
            self.belief.mean += K.dot(z - hmu)
            self.belief.cov = (np.eye(3) - K.dot(H)).dot(self.belief.cov)
            
            
    #移動更新
    def motion_update(self, nu, omega, time): 
        if abs(omega) < 1e-5: omega = 1e-5 #ゼロにすると式が変わるので避ける
        if abs(nu) < 1e-5:         nu = 1e-5

        v = self.motion_noise_stds##ノイズのパラメーターをvに代入
        M = np.diag([v["nn"]**2*abs(nu)/time + v["no"]**2*abs(omega)/time, 
                     v["on"]**2*abs(nu)/time + v["oo"]**2*abs(omega)/time])##ノイズの共分散行列Mの計算
        
        ##Aの計算
        t = self.belief.mean[2]##θ確率密度分布の平均値をtに代入
        st, ct = math.sin(t), math.cos(t)##tのsin,cosを計算 ロボットの進行方向の速度をx,y軸に振り分ける
        stw, ctw = math.sin(t + omega*time), math.cos(t + omega*time)##サンプリングタイム間で変化した進行速度の計算
        A = np.array([[(stw - st)/omega,    -nu/(omega**2)*(stw - st) + nu/omega*time*ctw],
                                 [(-ctw + ct)/omega, -nu/(omega**2)*(-ctw + ct) + nu/omega*time*stw],
                                 [0,                                time]] )##Aの計算
        
        #Fの計算　Fを使い分布の中心からずれた位置での状態遷移（f(μt−1,ut)）を補正する　（テイラー展開で一次関数に近似）
        F = np.diag([1.0, 1.0, 1.0])##3ｘ3の段位行列の生成
        F[0, 2] = nu / omega * (math.cos(t + omega * time) - math.cos(t))
        F[1, 2] = nu / omega * (math.sin(t + omega * time) - math.sin(t))
        
        #移動後の確率密度分布の平均値と共分散行列を計算　（更新）
        self.belief.cov = F.dot(self.belief.cov).dot(F.T) + A.dot(M).dot(A.T)##共分散行列
        self.belief.mean = IdealRobot.state_transition(nu, omega, time, self.belief.mean)##平均
        
    def draw(self, ax, elems):
        ###xy平面上の誤差の3シグマ範囲###
        e = sigma_ellipse(self.belief.mean[0:2], self.belief.cov[0:2, 0:2], 3)
        elems.append(ax.add_patch(e))

        ###θ方向の誤差の3シグマ範囲###
        x, y, c = self.belief.mean
        sigma3 = math.sqrt(self.belief.cov[2, 2])*3
        xs = [x + math.cos(c-sigma3), x, x + math.cos(c+sigma3)]##描画する２つの線分の端点のx座標
        ys = [y + math.sin(c-sigma3), y, y + math.sin(c+sigma3)]##描画する２つの線分の端点のy座標
        elems += ax.plot(xs, ys, color="blue", alpha=0.5)
```

```python
class KfAgent(Agent): 
    def __init__(self, time_interval, nu, omega, init_pose, envmap, \
                motion_noise_stds={"nn":0.19, "no":0.001, "on":0.13, "oo":0.2}):#ノイズのパラメーター
        super().__init__(nu, omega)
        self.kf = KalmanFilter(envmap, init_pose, motion_noise_stds)#カルマンフィルター
        self.time_interval = time_interval##サンプリングタイム
        
        self.prev_nu = 0.0
        self.prev_omega = 0.0
        
    def decision(self, observation=None):  ###kfagent2
        self.kf.motion_update(self.prev_nu, self.prev_omega, self.time_interval)#移動更新
        self.prev_nu, self.prev_omega = self.nu, self.omega#更新後の信念の平均値，分散を代入
        self.kf.observation_update(observation)#計測更新
        return self.nu, self.omega
        
    def draw(self, ax, elems): #追加
        self.kf.draw(ax, elems)
```
